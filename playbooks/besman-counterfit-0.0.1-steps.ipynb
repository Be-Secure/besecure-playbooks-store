## Counterfit Guide
This document provides the step by step guide to get the attack results on a AI model under asessment.

### Pre-Requisite

The model file, input file and target file should be under ~/counterfit/counterfit/targets. If the model and corresponding files are not present, follow the instructions here to [onboard an existing model](https://github.com/Be-Secure/counterfit/tree/develop?tab=readme-ov-file#onboard-existing-targets). If the user want to create a new target(model) then, follow the instructions [here](https://github.com/Be-Secure/counterfit/tree/develop?tab=readme-ov-file#onboard-new-target).


### Steps
#### 1. Change directory to counterfit
`cd counterfit`

#### 2. Activate conda environment.
`conda activate counterfit`

#### 3. Activate counterfit.
`counterfit`

verify that the prompt changes to \(coounterfit>\)

#### 4. List available targets
`list targets`

Check the target to be accessed is present in list.

#### 5. Add new target. \(This step is applicable only if the target is not available in list\) 
`new -n <new model name i.e BESMAN_ARTIFACT_NAME parameter> -d image`

 The new target name should be same as BESMAN_ARTIFACT_NAME and should have been git cloned. 

#### 6. Copy the model files to counterfit target.  \(This step is applicable only if the target is not available in list\) 
`cp -f <BESMAN_ARTIFCAT_NAME>/counterfit/<BESMAN_ARTIFCAT_NAME>.py counterfit/targets/<BESMAN_ARTIFCAT_NAME>.py`
`cp -f <BESMAN_ARTIFCAT_NAME>/counterfit/<BESMAN_ARTIFCAT_NAME>.npz counterfit/targets/<BESMAN_ARTIFCAT_NAME>/<BESMAN_ARTIFCAT_NAME>.npz`
`cp -f <BESMAN_ARTIFCAT_NAME>/<BESMAN_ARTIFCAT_NAME>.h5 counterfit/targets/<BESMAN_ARTIFCAT_NAME>/<BESMAN_ARTIFCAT_NAME>.h5`

#### 7. List available targets
`list targets`

Check the new target is present in list.

#### 8. Set the target
`set_target <target_name>`

#### 9. List the availbale attacks
`list attacks`

verify that the required attack is available in list.

#### 10. Set the required attack.
`set_attack <attack_name>`

This will display the attack id on the screen capture the attack id. We will need it later below.
e.g
\[+\] success: Using \<attackid\>

#### 11. Test model without attack.
`predict -i <sample_index>`

sample_index is the index of file given as input to the model under test. The model is provided with a set of input bundled in a .npz file (This file is expected to be created when we the tartget folder under counterfit is created). Based on the index number in this command the input will be feed to the model for prediction.

Verify that the model predicts the output as expected.

#### 12. Now set the model prameters for attack.
`set_params <list of params and their value in format \"--\<key\> value\">`

e.g set_params --sample_index=5 --max_eval 5000 --max_iter 10

To know the parameters use `show options` command.

#### 13. Run the attcack
`run`

#### 14. Check the model output now 
`predict -a`

#### 15. Save the results
`save -r`

#### 16. Exit from conda
`Exit counterfit`

#### 17. Deactivate Conda 
`conda deactivate`

#### 18. Export the attack id captured above
`echo "export COUNTERFIT_ATTACKID=<ID captured above>" >> ~/.bashrc`
`source ~/.bashrc`
